```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Activation, BatchNormalization, Dropout, ZeroPadding2D
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model

# Define the CNN model
model = Sequential([
    Conv2D(64, (9, 9), input_shape=(100, 100, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (5, 5), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    ZeroPadding2D((1, 1)),
    Conv2D(256, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    ZeroPadding2D((1, 1)),
    Conv2D(512, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),

    ZeroPadding2D((1, 1)),
    Conv2D(512, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(64),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.5),

    Dense(128),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.5),

    Dense(1),
    BatchNormalization(),
    Activation('sigmoid')
])

# Plot the model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)

# Display the model architecture
plt.figure(figsize=(10, 10))
img = plt.imread('model_plot.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```
